```python
import numpy as np
import pandas as pd
import sklearn as sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
diabetes_dataset=pd.read_csv ('Downloads/diabetes.csv')
```

```python
diabetes_dataset.head (4)
```

```python
diabetes_dataset.shape
```

```python
# separate dataset and labels

X=diabetes_dataset.drop(columns='Outcome', axis=1)
Y=diabetes_dataset['Outcome']
print(X)
```

```python
print(Y)
```

```python
# data standardisation
scaler=StandardScaler()
scaler.fit(X)
standardised_data=scaler.transform(X)
print(standardised_data)
```

```python
X=standardised_data
Y=diabetes_dataset['Outcome']
```

```python
#train/test split of data
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2, stratify=Y, random_state=2)
```

```python
# training the data
classifier=svm.SVC(kernel='linear')

#  training SVM
classifier.fit(X_train, Y_train)     
```

```python
#accuracy scoX_trainin
X_train_prediction=classifier.predict (X_train)
training_data_accuracy=accuracy_score(X_train_prediction, Y_train)
print('train data accuracy is :',training_data_accuracy)
```

```python
# accuracy_test data
X_test_prediction=classifier.predict (X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
print('train data accuracy is :',test_data_accuracy)
```

```python
# making predictions
input_data=(83,78,50,32,88,31,0.248,26)
#change data as np
input_data_as_np=np.asarray(input_data)
# reshape the np array data
input_data_reshaped=input_data_as_np.reshape(1,-1)
# standardise the input data
std_data=scaler.transform(input_data_reshaped)
print(std_data)
```

```python
prediction=classifier.predict(std_data)
print(prediction)
```
