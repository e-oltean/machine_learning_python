```python
# import libraries

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
# load datasat
data=pd.read_csv('Downloads/Decision_Tree_ Dataset.csv')
```

```python
data.head
```

```python
#data.describe
```

```python
# separation of features/target
x=data.iloc[:,1:5]
y=data.iloc[:,-1]
```

```python
# split the data into training and test data
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.33,random_state=4)
```

```python
clsf=DecisionTreeClassifier(criterion="gini", random_state=4, max_depth=3, min_samples_leaf=5)
```

```python
clsf.fit(x_train,y_train)
```

```python
#prediction 
y_predicted=clsf.predict(x_test)
```

```python
print(y_predicted)
```

```python
#verify accuracy
accuracy=accuracy_score(y_test, y_predicted)
print(accuracy)
```
